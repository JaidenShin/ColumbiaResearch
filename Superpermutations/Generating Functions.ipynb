```python
import numpy as np
import math
```

```python
def binomial(a,b):
  return math.factorial(a) / (math.factorial(b) * math.factorial(a-b))
```

```python
def findProb(n,m):
  prob = 0.0
  start = m//n
  while (start-m) % (n-1) != 0 or start*n < m:
    start += 1
  print(start)
  for k in range(start, m+1, n-1):
    j = int((n*k-m)/(n-1))
    prob += binomial(k,j) * (-1/n)**(n*(k-j))
  return prob
```

```python
print(findProb(7,1000))
```
