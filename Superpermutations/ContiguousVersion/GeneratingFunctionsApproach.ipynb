```python
# Generating Functions based approach to superpermutation length
# We consider a generating function that represents number of strings
# that do not contain a specific permutation, and use a union-bound approach.

import numpy as np
import math
```

```python
binom = [[0 for j in range(6000)] for i in range(6000)]
```

```python
for a in range(1,6000):
  for b in range(1,a+1):
    if b == 1:
      binom[a][b] = a
    else:
      binom[a][b] = binom[a-1][b] + binom[a-1][b-1]
```

```python
# setting up p(x)

n = 4
maxCheck = 35
deg = 33

pArr = np.zeros(n-1)
for i in range(n-1):
  pArr[i] = int(math.factorial(n) / math.factorial(i+2))

p = np.poly1d(pArr)
print(p)
```

```python
aArr = np.zeros(n)
aArr[0] = int(math.factorial(n)/2)

top = p * np.poly1d([-3,1]) + np.poly1d(aArr)

rArr = np.zeros(2*n)
rArr[0] = int((-1) * math.factorial(n)/2)
rArr[2*n-2] = 3
a = np.poly1d([-3,1])
bArr = np.zeros(n+1)
bArr[0] = 1
b = np.poly1d(bArr)
r = np.poly1d(rArr) - b * a * p

finalPoly = p * np.poly1d([-3,1]) + np.poly1d(aArr)
for i in range(1,maxCheck+1):
  finalPoly +=  top * (r ** i)

print(finalPoly)
```

```python
coefs = finalPoly.c
print(2 * coefs[len(coefs)-1-deg] / (3**(deg-n+2)))
```
