```python
import numpy as np
import math
```

```python
def analyze(N, L):
  arr = np.zeros(L)
  for i in range(L):
    arr[i] = int(np.random.uniform(0,N))
  alphas = np.zeros(L)
  count = np.zeros(N)
  perm = True
  for i in range(L):
    for start in range(max(0,i-N+1), min(L,i+N)):
      for j in range(start, min(start+N,L)):
        count[int(arr[j])] += 1
      perm = True
      for k in range(N):
        if count[k] != 1:
          perm = False
        count[k] = 0
      if perm:
        alphas[i] += 1
  print(alphas)
```

```python
analyze(4,10)
```
