```python
import numpy as np
import math
import matplotlib.pyplot as plt

R = 1
L = 1
N = 10**2
T = 10**6
```

```python
numLine = np.linspace(0, 2*N, 2*N+1)
waitingTimes = np.zeros(2*N-1)
posN = np.zeros(T+1)
times = np.zeros(T+1)
timeElapsed = 0;
currNPosition = N
posN[0] = N
for i in range(2*N-1):
  waitingTimes[i] = (-1/R) * np.log(np.random.uniform(0,1)) if numLine[i] < numLine[i+1] else (-1/L) * np.log(np.random.uniform(0,1))

for t in range(1,T+1):
  if t % 1000 == 0:
    print(t)
  minPosition = 0
  minWT = waitingTimes[0]
  for i in range(2*N-1):
    if minWT > waitingTimes[i]:
      minPosition = i
      minWT = waitingTimes[i]
  for i in range(2*N-1):
    if i == minPosition:
      temp = numLine[i]
      numLine[i] = numLine[i+1]
      numLine[i+1] = temp
      if numLine[i] == N:
        currNPosition = i
      if numLine[i+1] == N:
        currNPosition = i+1
      waitingTimes[i-1] = (-1/R) * np.log(np.random.uniform(0,1)) if numLine[i-1] < numLine[i] else (-1/L) * np.log(np.random.uniform(0,1))
      waitingTimes[i] = (-1/R) * np.log(np.random.uniform(0,1)) if numLine[i] < numLine[i+1] else (-1/L) * np.log(np.random.uniform(0,1))
      if i+1 < 2*N-1:
        waitingTimes[i+1] = (-1/R) * np.log(np.random.uniform(0,1)) if numLine[i+1] < numLine[i+2] else (-1/L) * np.log(np.random.uniform(0,1))
    elif i != minPosition-1 and i != minPosition+1:
      waitingTimes[i] -= minWT
  timeElapsed += minWT
  times[t] = timeElapsed
  posN[t] = currNPosition

plt.plot(times, posN)
plt.show();
```
