```python
import numpy as np
import matplotlib.pyplot as plt
import statistics as stats
```

```python
#graph last passage percolation times divided by N, in order to find the constant C by which N is scaled to be used as an offset. Numerical estimates find C ~= 3.92, which matches the theortically value up to bounds by the central limit theorem. 


def findingN(N,M):
  #gumbel dsitirbuted edge weigths, H = horizantal, V = vertical, computed through the quantile function
  H, V = -np.log(-np.log(np.random.random((M + 1, N + 1)))), -np.log(-np.log(np.random.random((M + 1, N + 1)))) 
  T = np.zeros((M + 1, N + 1))
  
  
  #x = M, y = N
  #simulate last passage percolation time
  T[0][0] = 0
  for i in range(M + 1):
    for j in range(N + 1):
      if (i != 0 and j != 0):
        T[i][j] = max(T[i-1][j] + H[i][j], T[i][j-1] + V[i][j])
      elif (i != 0 and j == 0):
        T[i][j] = T[i-1][j] + H[i][j]
      elif (i == 0 and j != 0):
        T[i][j] = T[i][j-1] + V[i][j]
  

  #graph last passage peroclation times to find C
  TNValues = np.zeros(N)
  for i in range(N):
    TNValues[i] = T[i][i] / i
  x = np.linspace(1,N,N)
  y = np.full(N,3.927)
  y2 = np.full(N,4)
  print(T[N][N] / N)
  plt.plot(x, TNValues)
  plt.plot(x, y, color = "r")
  plt.plot(x, y2, color = "g")

  plt.show()
```

```python
findingN(10**4, 10**4)
```
